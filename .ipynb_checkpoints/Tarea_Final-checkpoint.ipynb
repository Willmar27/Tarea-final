Lo primero es importar las librerías que se utilizarán para la resolución de la tarea.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import curve_fit
```

### Puntos 1 al 3:
Primero se leen los archivos .csv que contienen los datos.

```python
dat1=pd.read_csv("data/Global_active_power.csv")
dat2=pd.read_csv("data/Global_intensity.csv")
dat3=pd.read_csv("data/Global_reactive_power.csv")
dat4=pd.read_csv("data/Voltage.csv")
```

A continuación se fusionan todos los datos anteriores (dat1-4) en una única tabla de datos. Esto está en escala de minutos.

```python
data = pd.merge(dat1, dat2, on=['Datetime'])
data = pd.merge(data, dat3, on=['Datetime'])
data = pd.merge(data, dat4, on=['Datetime'])
data
```

Se realiza el resample de los datos para tenerlos en escalas de días y meses.

```python
data = data.set_index('Datetime')
data.index = pd.to_datetime(data.index)
datam = data.resample('M').mean()
datad = data.resample('D').mean()
```

### Punto 4 y 5

Se realizan las funciones para los ajustes. Téngase en cuenta que para la primera y segunda relación se usará la misma funcion 1. La función 2 se utilizará para la relación 3 correspondiente a una regresión múltiple.

```python
def func1(x, a, b): 
    return a*x+b   #Relaciones 1 y 2 propuestas en la tarea.

def func2(X, a1, a2, b): 
    x1,x2 = X
    return a1*x1 + a2*x2 + b #Relación 3 propuesta en la tarea.
```

El resample de los datos en días arrojaba un error al momento de realizar los ajustes porque habían valores NaN, con las siguientes líneas eso se solucionó:

```python
datad['Global_active_power'] = np.nan_to_num(datad['Global_active_power'])
datad['Global_reactive_power'] = np.nan_to_num(datad['Global_reactive_power'])
datad['Global_intensity'] = np.nan_to_num(datad['Global_intensity'])
datad['Voltage'] = np.nan_to_num(datad['Voltage'])

np.isnan(datad['Global_intensity']).any()
```

Se crea cierta funcion que toma la funcion 1 (func1) definida anteriormente, esta función (get_fit) tomará los datos (dependiendo de la escala de tiempo) y creara los ajustes. Esta función se utilizará para los ajustes 1 y 2.

```python
#Función tomada del archivo compartido por el profesor, esta se modificó un poco.
def get_fit(func,xdata,ydata,xdatad,ydatad,xdatam,ydatam,labelx,labely,r): 

#Con las siguientes líneas se realiza el ajuste y se hallan tanto los valores de los coeficientes de la función como errores estándar.
    popt1, pcov1 = curve_fit(func, xdata, ydata)
    popt1, pcov1 = np.around(popt1,r), np.around(np.sqrt(np.diag(pcov1)),r)
    
    popt2, pcov2 = curve_fit(func, xdatad, ydatad)
    popt2, pcov2 = np.around(popt2,r), np.around(np.sqrt(np.diag(pcov2)),r)
    
    popt3, pcov3 = curve_fit(func, xdatam, ydatam)
    popt3, pcov3 = np.around(popt3,r), np.around(np.sqrt(np.diag(pcov3)),r)
    
#A continuación se realiza lo necesario para los plots.    
    x=np.linspace(xdata.min(),xdata.max(),100)
    xd=np.linspace(xdatad.min(),xdatad.max(),100)
    xm=np.linspace(xdatam.min(),xdatam.max(),100)
    
#Se crean 3 subplots correspondientes a minutos, meses y días.
#Se sobreponen las gráficas de scatter y las rectas de ajuste en las 3 escalas (minutos, días, meses). Finalmente se grafica.
    plt.subplots(figsize=(16, 16))
    
    plt.subplot(3,2,1)
    plt.plot(x, func(x, *popt1), 'r-',label=r'Ajuste: {}$\pm${}'.format(np.around(tuple(popt1),r),tuple(pcov1)))
    plt.scatter(xdata,ydata,label="Datos originales")
    plt.xlabel(labelx)
    plt.ylabel(labely)
    plt.title("Ajuste para datos en escala de minutos")
    plt.legend()

    plt.subplot(3,2,2)
    plt.plot(xd, func(xd, *popt2), 'r-',label=r'Ajuste: {}$\pm${}'.format(np.around(tuple(popt2),r),tuple(pcov2)))
    plt.scatter(xdatad,ydatad,label="Datos originales")
    plt.xlabel(labelx)
    plt.ylabel(labely)
    plt.title("Ajuste para datos en escala de días")
    plt.legend()
    
    plt.subplot(3,2,3)
    plt.plot(xm, func(xm, *popt3), 'r-',label=r'Ajuste: {}$\pm${}'.format(np.around(tuple(popt3),r),tuple(pcov3)))
    plt.scatter(xdatam,ydatam,label="Datos originales")
    plt.xlabel(labelx)
    plt.ylabel(labely)
    plt.title("Ajuste para datos en escala de meses")
    plt.legend()
    
#Se imprimen en pantalla los coeficientes y errores del ajuste.
    plt.show()
    print("Los coeficientes para los datos en minutos son: {}".format(popt1))
    print("Los coeficientes sigma (errores estándar) son:: {}".format(pcov1),"\n")

    print("Los coeficientes para los datos en días son:  {}".format(popt2))
    print("Los coeficientes sigma (errores estándar) son: {}".format(pcov2),"\n")
    
    print("Los coeficientes para los datos en meses son: {}".format(popt3))
    print("Los coeficientes sigma (errores estándar) son: {}".format(pcov3),"\n")
    return #popt, np.sqrt(np.diag(pcov))
```

Ahora se construirá la función "get_fit2" que será utilizada para la tercera relación. Esta a diferencia de las dos anteriores depende de dos variables, por lo que se realizará una regresión lineal múltiple y se graficará su plano correspondiente.

```python
#Se creó una nueva función que recibe "Voltage", "Global Intensity" y "Global active power".
#Como son 3 casos (minutos, días y meses) entonces la función recibe estos 9 datos + el dato 'r' que sirve para tomar los decimales que se deseen.
def get_fit2(datintensity, datdintensity, datmintensity, datvoltage, datdvoltage, datmvoltage, datactive, datdactive, datmactive, r):
    
    #Con las siguientes lineas se usa curve_fit y la función 'func2' previamente creada y se realiza el ajuste.
    #Con esto se hallan los coeficientes y los errores estándar para almacenarlos en poptt y pcovv (uno por cada caso).
    poptt1, pcovv1= curve_fit(func2, (datintensity, datvoltage), datactive)
    poptt2, pcovv2= curve_fit(func2, (datdintensity, datdvoltage), datdactive)
    poptt3, pcovv3= curve_fit(func2, (datmintensity, datmvoltage), datmactive)
    
    #Se imprimen en pantalla los coeficientes y errores del ajuste.
    print("Los coeficientes para los datos en minutos son: {}".format(poptt1))
    print("Los coeficientes sigma (errores estándar) son:")
    print(pcovv1.diagonal(),"\n")
    
    print("Los coeficientes para los datos en días son: : {}".format(poptt2))
    print("Los coeficientes sigma (errores estándar) son:")
    print(pcovv2.diagonal(),"\n")
    
    print("Los coeficientes para los datos en meses son: : {}".format(poptt3))
    print("Los coeficientes sigma (errores estándar) son:")
    print(pcovv3.diagonal(),"\n")
    
    #Se comienza a realizar el código necesario para graficar, gran parte de esto lo construimos con prueba y error
    #mientras investigábamos en foros y sitios web como "Stack Overflow". Todo está hecho para los 3 casos (mins, días, meses)
    
    x1_range = np.arange(datintensity.min(), datintensity.max())
    y1_range = np.arange(datvoltage.min(), datvoltage.max())
    
    x2_range = np.arange(datdintensity.min(), datdintensity.max())
    y2_range = np.arange(datdvoltage.min(), datdvoltage.max())
    
    x3_range = np.arange(datmintensity.min(), datmintensity.max())
    y3_range = np.arange(datmvoltage.min(), datmvoltage.max())
    

    X1, Y1 = np.meshgrid(x1_range, y1_range)
    X2, Y2 = np.meshgrid(x2_range, y2_range)
    X3, Y3 = np.meshgrid(x3_range, y3_range)

    # Se utilizan los valores calculados del ajute con curve_fit y se crean las variables del eje Z para el ajuste
    
    Z1 = poptt1[2:3] + X1*poptt1[0:1] + Y1*poptt1[1:2]
    Z2 = poptt2[2:3] + X2*poptt2[0:1] + Y2*poptt2[1:2]
    Z3 = poptt3[2:3] + X3*poptt3[0:1] + Y3*poptt3[1:2]

    fig = plt.figure(figsize=(12,32)) # (largo,alto)
    
    #Se comienza a graficar en 3D añadiendo subplots, uno por cada caso. El plano de ajuste se sobrepone a los datos 
    #originales del scatter.
        
    ax = fig.add_subplot(3, 1, 1, projection='3d')
    ax.plot_surface(X1, Y1, Z1, rstride=1, cstride=1, alpha = 0.2)
    ax.scatter(datintensity, datvoltage, datactive, color='r', label="Datos originales")
    ax.view_init(10, 120)
    ax.dist = 10
    ax.set_xlabel('Global Intensity')
    ax.set_ylabel('Voltage')
    ax.set_zlabel('Global active power') 
    ax.text2D(0.24, 0.93, r'Plano de ajuste: {}$\pm${}'.format(np.around(tuple(poptt1),r), np.around(tuple(pcovv1.diagonal()),r)), transform=ax.transAxes,  color='b')
    ax.text2D(-0.03, 0.063, 'AJUSTE PARA DATOS EN ESCALA DE MINUTOS', color="black")
    plt.legend()
    
    ax = fig.add_subplot(3, 1, 2, projection='3d')
    ax.plot_surface(X2, Y2, Z2, rstride=1, cstride=1, alpha = 0.2)
    ax.scatter(datdintensity, datdvoltage, datdactive, color='r', label="Datos originales")
    ax.view_init(10, 120)
    ax.dist = 10
    ax.set_xlabel('Global Intensity')
    ax.set_ylabel('Voltage')
    ax.set_zlabel('Global active power') 
    ax.text2D(0.14, 0.93, r'Plano de ajuste: {}$\pm${}'.format(np.around(tuple(poptt2),r), np.around(tuple(pcovv2.diagonal()),r)), transform=ax.transAxes,  color='b')
    ax.text2D(-0.03, 0.063, 'AJUSTE PARA DATOS EN ESCALA DE DÍAS', color="black")
    plt.legend()

    ax = fig.add_subplot(3, 1, 3, projection='3d')
    ax.plot_surface(X3, Y3, Z3, rstride=1, cstride=1, alpha = 0.2)
    ax.scatter(datmintensity, datmvoltage, datmactive, color='r', label="Datos originales")
    ax.view_init(10, 120)
    ax.dist = 10
    ax.set_xlabel('Global Intensity')
    ax.set_ylabel('Voltage')
    ax.set_zlabel('Global active power') 
    ax.text2D(0.15, 0.93, r'Plano de ajuste: {}$\pm${}'.format(np.around(tuple(poptt3),r), np.around(tuple(pcovv3.diagonal()),r)), transform=ax.transAxes,  color='b')
    ax.text2D(-0.03, 0.063, 'AJUSTE PARA DATOS EN ESCALA DE MESES', color="black")
    plt.legend()
    return

    # En todas las lineas anteriores donde se construían los gráficos se nombraron los ejes, se cambió el ángulo de vista,
    # se añadieron las leyendas, entre otras cosas para mejorar la visualización.
```

##### A continuación se graficarán los datos originales (scatterplots) en sobreposición de sus gráficas de ajuste correspondientes para cada uno de los 3 ajustes y para cada caso (minutos, días, meses).

```python
get_fit2(data["Global_intensity"], datad["Global_intensity"], datam["Global_intensity"], data["Voltage"], datad["Voltage"], datam["Voltage"], data["Global_active_power"], datad["Global_active_power"], datam["Global_active_power"], 10)
```

En esta tercera relación en donde se involucran los tres datos, estando "Global intensity" y "Voltage" en el input del ajuste y "Global active power" en función de ellos, tiene como característica que la regresión lineal es múltiple (2 variables) y por lo tanto la mejor aproximación se trata de un plano.

Este ajuste múltiple representa mucho mejor los datos que una recta, ademas, tiene sigmas mucho mas pequeños que los ajustes anteriores (miles de veces mas pequeños).

Al igual que los anteriores casos de escala de tiempo, se observa un aumento del error conforme se aumenta la escala de entre 10 a 1000 veces en algunos casos. El ajuste óptimo se da en el caso de minutos.
